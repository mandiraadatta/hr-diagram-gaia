```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# The file has already been loaded into 'df' by the previous cell.
# Remove bad data
df = df.dropna(subset=['parallax','phot_g_mean_mag','bp_rp'])
df = df[df['parallax'] > 0]

# Calculate absolute magnitude
df['abs_magnitude'] = (
    df['phot_g_mean_mag']
    - 5*np.log10(df['distance_pc'])
    + 5
)

# Plot H-R Diagram
plt.figure(figsize=(10,12))

plt.scatter(
    df['bp_rp'],
    df['abs_magnitude'],
    s=0.5,
    alpha=0.3,
    c=df['bp_rp'],
    cmap='RdYlBu_r'
)

plt.gca().invert_yaxis()

plt.xlabel('BP-RP Color Index')
plt.ylabel('Absolute Magnitude')
plt.title('Hertzsprung-Russell Diagram')

plt.colorbar(label='BP-RP')
plt.show()
```
